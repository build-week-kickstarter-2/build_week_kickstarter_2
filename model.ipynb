```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
df
```

```python
df.drop(columns=['ID','name','category','currency','pledged','backers','usd pledged','usd_pledged_real','goal'], inplace=True)
```

```python

df = df.drop(df.query('''country == 'N,0"' ''').index, axis=0).reset_index(drop=True)
```

```python
df['country'].value_counts()
```

```python
df['country'] = df['country'].map({'US':1, 
                                    'GB':2,
                                    'CA':3,
                                    'AU':4,
                                    'DE':5,
                                    'FR':6,
                                    'NL':7,
                                    'IT':8,
                                    'ES':9,
                                    'SE':10,
                                    'MX':11,
                                    'NZ':12,
                                    'DK':13,
                                    'IE':14, 
                                    'CH':15,
                                    'NO':16,
                                    'BE':17,
                                    'AT':18,
                                    'HK':19,
                                    'SG':20,
                                    'LU':21,
                                    'JP':22
                                        })
```

```python
df
```

```python
df['main_category'] = df['main_category'].map({'Publishing':1, 
                                                'Film & Video':2,
                                                'Music':3,
                                                'Food':4,
                                                'Crafts':5,
                                                'Games':6,
                                                'Design':7,
                                                'Comics':8,
                                                'Fashion':9,
                                                'Theater':10,
                                                'Art':11,
                                                'Photography':12,
                                                'Technology':13,
                                                'Dance':14,
                                                'Journalism':15
                                                })
```

```python
df
```

```python
df = df.drop(df.query("state != 'failed' and state != 'successful'").index, axis=0).reset_index(drop=True)
```

```python
df
```

```python
df['state'] = df['state'].map({'successful':1, 'failed':0})
```

```python
df
```

```python
df['deadline'][0][:4]
df['deadline'][0][5:7]
df['deadline'][0][8:]
```

```python
df[['deadline','launched']] = df[['deadline','launched']].apply(pd.to_datetime)
df['days'] = (df['deadline'] - df['launched']).dt.days
```

```python
df



```

```python
df.drop(columns=['deadline','launched'],inplace=True)
```

```python
X = df.drop(columns='state')
y = df['state']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.2, random_state = 1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV


ks = [5, 10, 20, 25, 30,35]
param_grid = [{'n_neighbors': ks}]
knn = KNeighborsClassifier(n_neighbors=35)
knn_grid = GridSearchCV(knn, param_grid, cv=8, scoring='accuracy', verbose=1, n_jobs=-1)
knn_grid.fit(X_train, y_train)
```

```python
knn_grid.best_score_
```

```python
knn_grid.predict(X_test)
```

```python
def predict(main_category, country, usd_goal, days):
    df = pd.DataFrame(
        columns=['main_category', 'country', 'usd_goal', 'days'],
        data=[[main_category, country, usd_goal, days,]]
    )

    y_pred = knn_grid.predict(df)[0]
    if y_pred == 0:
        return "You won't make your goal"
    else:
        return 'You might make your goal!'
```

```python
predict(12,1,1000,15)
```
